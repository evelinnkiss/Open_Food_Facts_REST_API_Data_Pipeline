2. Import Libraries

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

3. Call API


```python
url = "https://world.openfoodfacts.org/cgi/search.pl"
params = {
    "action": "process",
    "tagtype_0": "categories",
    "tag_contains_0": "contains",
    "tag_0": "snacks",
    "page_size": 50,    # Number of products to fetch
    "json": True
}

response = requests.get(url, params=params)
if response.status_code != 200:
    raise Exception(f"API call failed with status code: {response.status_code}")

data = response.json()
```

4. Convert JSON to dataframe

```python
products = data.get("products", [])
df = pd.json_normalize(products)
```

```python
print("Columns available in the dataset:")
print(df.columns.tolist())
```

5. Select relevant columns

```python
columns_to_keep = ['product_name', 'brands', 'categories', 'countries', 'nutrition_grades']
df = df[columns_to_keep]
df = df.dropna(subset=['product_name'])  # remove products without names
```

6. Save to CSV

```python
df.to_csv('openfoodfacts_products.csv', index=False)
print("Data saved to openfoodfacts_products.csv")
df.head()
```
